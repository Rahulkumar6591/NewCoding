Feature selection part

```python
movie.columns
```

```python
movies=movie[['index', 'title', 'genres', 'Description']]
```

```python
movies
```

```python
movies['tags'] = movies['Description']+movies['genres']
```

```python
movies
```

Remove the Description and Genres

```python
new_data=movies.drop(columns=['Description','genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=7850,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Lion King 3D"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].title)
```

```python
def recommand(movie):
  index=new_data[new_data['title']==movie].index[0]
  distance=sorted(list(enumerate(similarity[2])),reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].title)
```

```python
recommand('The Lion King')
```
